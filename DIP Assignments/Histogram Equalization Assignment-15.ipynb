```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hist_plot(img):
    # list to store the count of each intensity value
    count = []
    # list to store intensity value
    r = []
    # loop to traverse each intensity value
    for k in range(0, 256):
        r.append(k)
        count1 = 0
        # loops to traverse each pixel in the image
        for i in range(m):
            for j in range(n):
                if img[i, j] == k:
                    count1 += 1
        count.append(count1)
    return (r, count)
```

```python
img = cv2.imread('flowers.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
m, n = img.shape


# calculating the histogram of the original image
r1, count1 = hist_plot(img)

# Applying histogram equalization
img1 = cv2.equalizeHist(img)
r2, count2 = hist_plot(img1)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title("Original Image")


plt.subplot(2, 2, 2)
plt.stem(r1, count1)
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.title('Histogram of the original image')

plt.subplot(2, 2, 3)
plt.imshow(img1)
plt.title("Equalized Image")


plt.subplot(2, 2, 4)
plt.stem(r2, count2)
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.title('Histogram of the equalized image')

plt.show()
```
