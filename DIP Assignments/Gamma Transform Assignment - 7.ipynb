```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('flowers.jpg')
```

```python
# gamma values to apply
gamma_values = [0.5, 1.0, 1.5, 2.0, 2.5]

# lookup table for gamma value
lookup_tables = [np.uint8(255 * np.power(np.arange(0, 256) / 255.0, gamma)) for gamma in gamma_values]

# gamma transformation to input image with lookup table
gamma_images = [cv2.LUT(img, table) for table in lookup_tables]
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
for i in range(len(gamma_values)):
    plt.subplot(2, 3, i+2)
    plt.imshow(cv2.cvtColor(gamma_images[i], cv2.COLOR_BGR2RGB))
    plt.title('Gamma = {}'.format(gamma_values[i]))

plt.show()
```
