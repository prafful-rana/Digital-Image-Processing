```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('flowers.jpg',1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
b, g, r = cv2.split(img)
```

```python
kernel_3 = np.ones((3, 3), np.float32)
kernel_5 = np.ones((5, 5), np.float32)


epsilon = 1e-5
gray_hm_3 = cv2.filter2D(gray, -1, 1/(cv2.filter2D(1/(gray+epsilon), -1, kernel_3) + epsilon))
gray_hm_5 = cv2.filter2D(gray, -1, 1/(cv2.filter2D(1/(gray+epsilon), -1, kernel_5) + epsilon))


img_r_hm_3 = cv2.filter2D(r, -1, 1/(cv2.filter2D(1/(r + epsilon), -1, kernel_3) + epsilon))
img_g_hm_3 = cv2.filter2D(g, -1, 1/(cv2.filter2D(1/(g + epsilon), -1, kernel_3) + epsilon))
img_b_hm_3 = cv2.filter2D(b, -1, 1/(cv2.filter2D(1/(b + epsilon), -1, kernel_3) + epsilon))

img_r_hm_5 = cv2.filter2D(r, -1, 1/(cv2.filter2D(1/(r + epsilon), -1, kernel_5) + epsilon))
img_g_hm_5 = cv2.filter2D(g, -1, 1/(cv2.filter2D(1/(g + epsilon), -1, kernel_5) + epsilon))
img_b_hm_5 = cv2.filter2D(b, -1, 1/(cv2.filter2D(1/(b + epsilon), -1, kernel_5) + epsilon))
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(2,2,3)
plt.imshow(gray_hm_3, cmap='gray')
plt.title('Grayscale Image 3x3')

plt.subplot(2,2,4)
plt.imshow(gray_hm_5, cmap='gray')
plt.title('Grayscale Image 5x5')
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')

plt.subplot(2,2,3)
plt.imshow(img_r_hm_3, cmap='Reds')
plt.title('Red Channel 3x3')

plt.subplot(2,2,4)
plt.imshow(img_r_hm_5, cmap='Reds')
plt.title('Red Channel 5x5')
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')

plt.subplot(2,2,3)
plt.imshow(img_g_hm_3, cmap='Greens')
plt.title('Green Channel 3x3')

plt.subplot(2,2,4)
plt.imshow(img_g_hm_5, cmap='Greens')
plt.title('Green Channel 5x5')
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')

plt.subplot(2,2,3)
plt.imshow(img_b_hm_3, cmap='Blues')
plt.title('Blue Channel 3x3')

plt.subplot(2,2,4)
plt.imshow(img_b_hm_5, cmap='Blues')
plt.title('Blue Channel 5x5')

plt.show()
```
