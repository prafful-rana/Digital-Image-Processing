```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('flowers.jpg', 1)
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
b, g, r = cv2.split(img)
```

```python
gray_filtered_3 = cv2.medianBlur(gray_image, 3)
r_filtered_3 = cv2.medianBlur(r, 3)
g_filtered_3 = cv2.medianBlur(g, 3)
b_filtered_3 = cv2.medianBlur(b, 3)


gray_filtered_5 = cv2.medianBlur(gray_image, 5)
r_filtered_5 = cv2.medianBlur(r, 5)
g_filtered_5 = cv2.medianBlur(g, 5)
b_filtered_5 = cv2.medianBlur(b, 5)
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(2, 2, 3)
plt.imshow(gray_filtered_3, cmap='gray')
plt.title('Grayscale Image 3x3')

plt.subplot(2, 2, 4)
plt.imshow(gray_filtered_5, cmap='gray')
plt.title('Grayscale Image 5x5')
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')

plt.subplot(2, 2, 3)
plt.imshow(r_filtered_3, cmap='Reds')
plt.title('Red Channel 3x3')

plt.subplot(2, 2, 4)
plt.imshow(r_filtered_5, cmap='Reds')
plt.title('Red Channel 5x5')
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')

plt.subplot(2, 2, 3)
plt.imshow(g_filtered_3, cmap='Greens')
plt.title('Green Channel 3x3')

plt.subplot(2, 2, 4)
plt.imshow(g_filtered_5, cmap='Greens')
plt.title('Green Channel 5x5')
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')

plt.subplot(2, 2, 3)
plt.imshow(b_filtered_3, cmap='Blues')
plt.title('Blue Channel 3x3')

plt.subplot(2, 2, 4)
plt.imshow(b_filtered_5, cmap='Blues')
plt.title('Blue Channel 5x5')

plt.show()
```
