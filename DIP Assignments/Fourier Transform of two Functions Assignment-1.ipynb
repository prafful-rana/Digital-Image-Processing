```python
import math
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0.01,1,5000)
fx=[]
for i in t:
    fx.append(math.sin(2*math.pi/i))
plt.plot(t,fx)
```

```python
t1 = np.linspace(0,1,1000)
length = 500
step = 500
val = 2
fx1=[]
for i in range(0,1000):
    if i < length:
        fx1.append(val)
    else:
        length=length+(step/2)
        step=step/2
        val=val/2
        fx1.append(val)
        
fx1.reverse()       
plt.plot(t1,fx1)
```

```python
from scipy.fft import fft
freq = fft(fx)
var = np.linspace(0,1,len(freq))
plt.plot(var,abs(freq))
```

```python
freq1 = fft(fx1)
var1 = np.linspace(0,1,len(freq1))
plt.plot(var1,abs(freq1))
```
