```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('flowers.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Apply log transformation to the grayscale image
log_transformed = np.log(1 + gray)

log_transformed = (255 * (log_transformed / np.max(log_transformed))).astype(np.uint8)
# cv2.normalize(log_transformed, log_transformed, 0, 255, cv2.NORM_MINMAX)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")

plt.subplot(2,2,2)
plt.hist(gray.ravel(), bins=256, range=(0, 256))
plt.title('Histogram of Original Image')
plt.xlim([0, 256])

plt.subplot(2,2,3)
plt.imshow(log_transformed, cmap='gray')
plt.title("Log Transformed Image")

plt.subplot(2,2,4)
plt.hist(log_transformed.ravel(), bins=256, range=(0, 256))
plt.title('Histogram of Log Transformed Image')
plt.xlim([0, 256])

plt.show()
```
