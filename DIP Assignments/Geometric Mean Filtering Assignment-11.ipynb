```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('flowers.jpg',1)

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

b, g, r = cv2.split(img)
epsilon = 1e-5
```

```python
gray_img_3x3 = np.uint8(np.exp(cv2.boxFilter(np.log(gray_img.astype('float32')+epsilon), -1, (3, 3))))
gray_img_5x5 = np.uint8(np.exp(cv2.boxFilter(np.log(gray_img.astype('float32')+epsilon), -1, (5, 5))))

# apply geometric mean filter to the red channel using the 3x3 kernel and 5x5 kernel
r_3x3 = np.uint8(np.exp(cv2.boxFilter(np.log(r.astype('float32')+epsilon), -1, (3, 3))))
r_5x5 = np.uint8(np.exp(cv2.boxFilter(np.log(r.astype('float32')+epsilon), -1, (5, 5))))

# apply geometric mean filter to the green channel using the 3x3 kernel
g_3x3 = np.uint8(np.exp(cv2.boxFilter(np.log(g.astype('float32')+epsilon), -1, (3, 3))))
g_5x5 = np.uint8(np.exp(cv2.boxFilter(np.log(g.astype('float32')+epsilon), -1, (5, 5))))

# apply geometric mean filter to the blue channel using the 3x3 kernel
b_3x3 = np.uint8(np.exp(cv2.boxFilter(np.log(b.astype('float32')+epsilon), -1, (3, 3))))
b_5x5 = np.uint8(np.exp(cv2.boxFilter(np.log(b.astype('float32')+epsilon), -1, (5, 5))))
```

```python
# plot the images
plt.figure(figsize=(15, 15))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(2,2,2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(2,2,3)
plt.imshow(g_3x3, cmap='gray')
plt.title('Grayscale Image 3x3')

plt.subplot(2,2,4)
plt.imshow(g_5x5, cmap='gray')
plt.title('Grayscale Image 5x5')
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(2,2, 1)
plt.imshow(img )
plt.title('Original Image')

plt.subplot(2,2, 2)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')

plt.subplot(2,2, 3)
plt.imshow(r_3x3, cmap='Reds')
plt.title('Red Channel 3x3')

plt.subplot(2,2, 4)
plt.imshow(r_5x5, cmap='Reds')
plt.title('Red Channel 5x5')
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(2,2, 1)
plt.imshow(img )
plt.title('Original Image')

plt.subplot(2,2, 2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')

plt.subplot(2,2, 3)
plt.imshow(g_3x3, cmap='Greens')
plt.title('Green Channel 3x3')

plt.subplot(2,2, 4)
plt.imshow(g_5x5, cmap='Greens')
plt.title('Green Channel 5x5')
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(2,2, 1)
plt.imshow(img )
plt.title('Original Image')

plt.subplot(2,2, 2)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')

plt.subplot(2,2, 3)
plt.imshow(b_3x3, cmap='Blues')
plt.title('Blue Channel 3x3')

plt.subplot(2,2, 4)
plt.imshow(b_5x5, cmap='Blues')
plt.title('Blue Channel 5x5')


plt.show()
```
